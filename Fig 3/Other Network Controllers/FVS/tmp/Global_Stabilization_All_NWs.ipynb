```python
import ast
import os
import json

# 경로 재지정
info_path = "00_Atractor_Major_Info2.txt"
network_folder = "./models2/"
output_path = "00_Atractor_Major_New.txt"

# 결과 리스트
converted_info = []
converted_info = []

# 한 줄씩 읽어서 처리
with open(info_path, "r") as f:
    lines = f.readlines()

for line in lines:
    try:
        # 네트워크 이름, info 문자열 분리
        network_name, info_str = line.strip().split(",", 1)
        info_str = info_str.strip()

        # JSON 파싱 시도
        try:
            info = json.loads(info_str)
        except json.JSONDecodeError:
            # JSON이 아닐 경우 ast로 대체 시도
            import ast
            info = ast.literal_eval(info_str)

        node_names = info["node_names"]
        state = info["attractor_state"]

        # 네트워크 경로
        net_path = os.path.join(network_folder, network_name.replace(".bnet", ".txt"))
        if not os.path.exists(net_path):
            print(f"File not found: {net_path}")
            continue

        # 네트워크 파일에서 노드 순서 읽기
        with open(net_path, "r") as f:
            net_lines = f.readlines()
        lhs_nodes = [line.split("=")[0].strip() for line in net_lines if "=" in line]

        # 노드명 → state 딕셔너리
        state_dict = dict(zip(node_names, state))

        # 새로운 순서로 state 재정렬
        reordered_state = "".join(state_dict[n] for n in lhs_nodes if n in state_dict)

        # info 업데이트
        info["node_names"] = lhs_nodes
        info["attractor_state"] = reordered_state

        # 다시 JSON 문자열로 저장
        converted_info.append(f"{network_name},{json.dumps(info)}\n")

    except Exception as e:
        print(f"Error in {network_name}: {e}")
        continue

# 파일로 저장
with open(output_path, "w") as f:
    f.writelines(converted_info)

output_path
```

```python
import re

def convert_node_name(name):
    if name in {"or", "and", "not"}:
        return name  # 논리 연산자는 그대로
    return (
        name.replace("or", "ox")
            .replace("and", "anx")
            .replace("not", "nox")
    )

def convert_logic_line(line):
    if "=" not in line:
        return line

    lhs, rhs = line.split("=", 1)
    lhs = lhs.strip()
    rhs = rhs.strip()

    # 좌변 치환
    lhs_converted = convert_node_name(lhs)

    # 우변: 단어 기준으로 나눠서 치환
    tokens = re.split(r'(\W)', rhs)  # 비단어 문자 기준 split (괄호, 공백 등 유지)
    tokens_converted = [convert_node_name(tok) if re.match(r'\w+', tok) else tok for tok in tokens]
    rhs_converted = ''.join(tokens_converted)

    return f"{lhs_converted} = {rhs_converted}\n"
```

```python
import glob
model_files = glob.glob("./models2/*.txt")

for file_path in model_files:
    with open(file_path, "r") as f:
        lines = f.readlines()
    converted_lines = [convert_logic_line(line) for line in lines]
    with open(file_path, "w") as f:
        f.writelines(converted_lines)
```

```python
model_paths = glob.glob("./models2/*.txt")
def rename_nodes_bracketed(lines):
    # 좌변 노드 수집
    node_order = [line.split("=")[0].strip() for line in lines if "=" in line]
    node_map = {orig: f"N_{orig}_N" for orig in node_order}

    renamed_lines = []

    for line in lines:
        if "=" not in line:
            renamed_lines.append(line)
            continue

        lhs, rhs = line.split("=", 1)
        lhs = lhs.strip()
        rhs = rhs.strip()

        # 우변 치환
        tokens = re.split(r'(\W)', rhs)
        tokens = [node_map.get(tok, tok) if re.match(r'\w+', tok) else tok for tok in tokens]
        rhs_renamed = ''.join(tokens)

        # 좌변 치환
        lhs_renamed = node_map.get(lhs, lhs)

        renamed_lines.append(f"{lhs_renamed} = {rhs_renamed}\n")

    return renamed_lines

# 전체 파일에 대해 적용 및 덮어쓰기
for path in model_paths:
    with open(path, "r") as f:
        lines = f.readlines()

    renamed_lines = rename_nodes_bracketed(lines)

    with open(path, "w") as f:
        f.writelines(renamed_lines)

len(model_paths)
```

```python
FVS_dict = {}         # 네트워크별 FVS 저장
error_indices = [] 

for i in range(36):
    try:
        with open(df['Results'][i], "r") as f:
            modeltext = f.read()
            
        modeltext = modeltext.replace("=", "*=").replace(" & ", " and ").replace(" | ", " or ").replace("~ ", "not ")
        modeltext = modeltext.strip()
        examNet = cana.boolean_network.BooleanNetwork.from_string_boolean(modeltext)
        #print(examNet)

        # Mapping nodes
        mappindDic = {}
        for node in examNet.nodes:
            mappindDic[node.id] = node.name
        # FVSs
        FVS_bruteforce = examNet.feedback_vertex_set_driver_nodes(graph='structural', method='bruteforce', max_search=10, keep_self_loops=True) # brutuforce
        #print(FVS_bruteforce)
        FVS_stack = []
        for FVS in FVS_bruteforce:
            FVS_list = []
            for node in FVS:
                FVS_list.append(mappindDic[node])
            FVS_stack.append(FVS_list)
        
        
        FVS_dict[df['Network'][i]] = FVS_stack
        
        
    except Exception as e:
        print(f"{i}에서 에러 발생: {e}")
        error_indices.append(i)

print("에러가 발생한 인덱스:", error_indices)
```
